### Q2

Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution.

```python
# Ans:- 
# Load the flight price dataset
df=pd.read_excel('flight_price.xlsx')
```

```python
# Create a histogram of flight prices
plt.hist(df['Price'], bins=20, edgecolor='black')
plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')
plt.show()
```

### Q3

Q3. What is the range of prices in the dataset? What is the minimum and maximum price?

```python
# Ans:- 
# the range of prices in the dataset and find the minimum and maximum price values
# Calculate the minimum price
min_price = df['Price'].min()

# Calculate the maximum price
max_price = df['Price'].max()

# Calculate the range of prices
price_range = max_price - min_price

# Display the results
print("Minimum Price:", min_price)
print("Maximum Price:", max_price)
print("Price Range:", price_range)
```

### Analyze Outliers:
After generating the boxplot, you can visually inspect it to identify any potential outliers. Outliers are data points that significantly deviate from the overall pattern of the data. In a boxplot, outliers are represented by individual points outside the whiskers of the box.

Outliers can impact your analysis in several ways:

They can skew statistical measures like the mean and standard deviation, as these measures are sensitive to extreme values.
Outliers may influence the results of certain analyses or models that assume a normal distribution or have specific assumptions about the data.
They can affect the accuracy and performance of predictive models by introducing noise or bias.
It's crucial to consider the nature of your analysis and the specific domain knowledge to determine how to handle outliers. Options include removing outliers, transforming the data, or using robust statistical methods that are less affected by extreme values.

### Q13

Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
Ans:- 
1. Import the necessary libraries:
2. Load the Google Playstore dataset:
3. Create the bar chart:
```

```python
filtered_df = df[['Type', 'Price']]
```

```python
average_prices = filtered_df.groupby('Type')['Price'].mean()
```

```python
plt.bar(average_prices.index, average_prices.values, color='skyblue')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.title('Average Prices by App Type')
plt.xticks(rotation=0)
plt.show()
```

Running these steps will calculate the average prices for each app type (Free or Paid) and create a bar chart comparing the average prices by app type. The app types will be shown on the x-axis, and the average prices will be displayed on the y-axis. The bar chart allows you to visually compare the average prices between different app types.

Note: Adjust the colors, axis labels, title, and rotation of the x-axis labels as per your preference for better visualization and readability.
