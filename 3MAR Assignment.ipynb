```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
```

```python
Ans:- Seaborn is a data visualization library built on top of Matplotlib. It provides a high-level interface 
for creating informative and attractive statistical graphics. Some of the plots that can be created using Seaborn 
    are:
```

```python
=> Scatter plot - This plot is used to visualize the relationship between two continuous variables. It helps to
identify the correlation between the variables.

=> Histogram - A histogram is a graphical representation of the distribution of a continuous numerical variable.
It helps to identify the underlying frequency distribution of the data.

=> Box plot - A box plot is used to show the distribution of data based on the five-number summary. It helps to 
identify the outliers and the quartile ranges.

=> Heatmap - A heatmap is a graphical representation of data where the values are represented as colors. It helps
to identify patterns and relationships in the data.

=> Violin plot - A violin plot is similar to a box plot but shows a more detailed representation of the data. 
It helps to identify the shape and density of the distribution of the data.
```

### Q2

```python
Q2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
```

```python
Ans:- the code to load the "fmri" dataset from Seaborn and plot a line plot using x = "timepoint" 
and y = "signal" for different events and regions:
```

```python
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset('fmri')

# Plot line plot for different events and regions
sns.lineplot(x='timepoint', y='signal', hue='event', style='region', data=fmri)
```

```python
This code will plot a line plot for the "fmri" dataset using the "timepoint" column as x-axis and the "signal" 
column as y-axis. The plot will be differentiated based on the "event" column using different colors, and
further differentiated based on the "region" column using different line styles.
```

### Q3

```python
Q3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
```

```python
Ans:- the code to load the "titanic" dataset using seaborn and plot two box plots using the boxplot() function:
```

```python
import seaborn as sns

# Load titanic dataset
titanic = sns.load_dataset('titanic')

# Create box plot for 'pclass' vs 'age'
sns.boxplot(x='pclass', y='age', data=titanic)

# Create box plot for 'fare'
sns.boxplot(y='fare', data=titanic)
```

```python
This will create two separate box plots. The first one will have 'pclass' on the x-axis and 'age' on the y-axis, 
while the second one will have 'fare' on the y-axis.
```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
```

```python
Ans:- the code to plot a histogram for the 'price' column of the 'diamonds' dataset using the 'cut' column as 
hue parameter:
```

```python
import seaborn as sns

diamonds = sns.load_dataset('diamonds')

sns.histplot(data=diamonds, x='price', hue='cut')
```

```python
This will generate a histogram of diamond prices with the different cuts represented by different colors.
The 'diamonds' dataset is included in seaborn and can be loaded using the load_dataset() function.
```

### Q6

```python
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
```

```python
Ans:- the code to plot a heatmap using the "flights" dataset from seaborn:
```

```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights = flights.pivot("month", "year", "passengers")
sns.heatmap(flights, cmap='coolwarm')
```

```python
This will create a heatmap of the passenger count for each month and year. The color scheme used here is 
'coolwarm', which goes from blue to red, with blue indicating low passenger counts and red indicating high
passenger counts.
```
