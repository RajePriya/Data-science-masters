### Q2

```python
Ans:- A scatter plot is a type of plot that displays values for two variables as a collection of points on a two
-dimensional plane. It is useful to identify patterns, trends, and outliers in the data.

To plot a scatter plot using the generated data, we can use the Pyplot module of Matplotlib 
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Q3

```python
Ans:- The subplot() function is used to create multiple plots in a single figure. It takes three arguments: the 
number of rows, the number of columns, and the plot number.
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with 2 rows and 2 columns
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot line 1 in the first subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')

# Plot line 2 in the second subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')

# Plot line 3 in the third subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

# Plot line 4 in the fourth subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')

plt.tight_layout()
plt.show()
```

### Q4

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
Ans:- A bar plot is a type of plot that represents categorical data with rectangular bars with heights or lengths 
proportional to the values that they represent. It is used to compare and visualize different categories of data.

To plot a bar plot and a horizontal bar plot using the given data, we can use the bar() function and barh() 
function respectively from the pyplot module of Matplotlib.
```

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title("Company Profits")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profits")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

```python
This will create a vertical bar plot and a horizontal bar plot, both displaying the company profits data.
```

### Q5

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
Ans:- A box plot is a graphical representation of a dataset showing the median, quartiles, and outliers of the 
data. It is used to display the distribution of a dataset and to identify any outliers or extreme values in the 
data.

To plot a box plot using the given data, we can use the boxplot() function from Matplotlib.
```

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()
ax.boxplot(data)

plt.title("Box plot of box1 and box2")
plt.show()
```

```python
This will generate a box plot of box1 and box2 with the title "Box plot of box1 and box2".
```
